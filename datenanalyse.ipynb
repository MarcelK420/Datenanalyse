## **Einleitung in das Thema**

In vielen Industriebranchen, zum Beispiel in der Automobil‑, Elektronik‑ oder Lebensmittelproduktion, werden Produkte in großen Stückzahlen hergestellt. Schon kleine Schwankungen im Produktionsprozess können zu Ausschuss, Nacharbeit oder Qualitätsproblemen beim Kunden führen, weshalb eine systematische und frühzeitige Überwachung der Prozessqualität notwendig ist.

Statistische Prozesskontrolle (Statistical Process Control, SPC) nutzt Methoden der Statistik, um Produktionsprozesse laufend anhand von Messdaten zu beobachten und zu steuern. Dabei werden Qualitätskennzahlen wie Mittelwerte, Spannweiten oder Ausschussanteile in sogenannten Kontrollkarten dargestellt, die neben einer Mittellinie auch obere und untere Eingriffsgrenzen enthalten; Überschreitungen dieser Grenzen oder charakteristische Muster deuten auf besondere Störeinflüsse hin und signalisieren, dass der Prozess nicht mehr stabil ist.

Ergänzend dazu ermöglichen Hypothesentests eine formale Bewertung von Prozessänderungen. Mit statistischen Tests wie dem t‑Test und der Varianzanalyse (Analysis of Variance, ANOVA) lässt sich prüfen, ob sich Kenngrößen eines Prozesses, etwa der Mittelwert einer Qualitätsmessgröße, nach einer Änderung tatsächlich signifikant verändert haben oder ob beobachtete Unterschiede auch durch zufällige Schwankungen erklärbar sind.

Ziel dieses Notebooks ist es, die grundlegenden Ideen von Hypothesentests (t‑Test und ANOVA) und statistischer Prozesskontrolle zu erläutern und anhand eines Datensatzes zu demonstrieren, wie sich beide Ansätze kombinieren lassen, um Prozessänderungen in der industriellen Qualitätssicherung zu bewerten.

## **Theoretische Grundlagen**
#### Mathematische und statistische Herleitung

Ein **Hypothesentest** prüft mit Hilfe einer Stichprobe eine Aussage über die Grundgesamtheit. Die **Nullhypothese $H_0$** beschreibt typischerweise "kein Unterschied", z.B. $ \mu_1 = \mu_2 $ (Mittelwerte zweier Prozesse sind gleich). Die **Alternativhypothese $H_1$** beschreibt den interessierenden Effekt, z.B. $ \mu_1 \neq \mu_2 $ [1].


Aus den Stichprobendaten wird eine **Teststatistik $T$** berechnet, die unter $H_0$ einer bekannten Verteilung folgt. Der **p-Wert** ist die Wahrscheinlichkeit, unter $H_0$ einen mindestens so extremen Wert beobachten: $$ p = P(T \geq t_{obs} | H_0) $$
Im folgenden wird ein Signifikanzniveau von $ \alpha = 0.05 $ verwendet, d.h. ein Fehler 1. Art von höchstens 5% wird akzeptiert:

Liegt $ p < \alpha $ (häufig $\alpha = 0{,}05$), wird $H_0 $ verworfen [1].

**t-Test** (Welch-t-Test für ungleiche Varianzen):

Für zwei unabhängige Stichproben $ X_1, \dots, X_{n_1}$ und $Y_1, \dots, Y_{n_2} $ mit den Mittelwerten $\bar{X}, \bar{Y}$ und Varianzen $s_X^2, s_Y^2$: $$ t = \frac{\bar{X} - \bar{Y}}{\sqrt{\frac{s_X^2}{n_1} + \frac{s_Y^2}{n_2}}} $$
$t$ folgt ungefähr einer t-Verteilung [1].

**ANOVA** (einfaktoriell):
Bei *k* Gruppen zerlegt ANOVA die Gesamtvarianz: $$ SS_gesamt = SS_B + SS_W $$
F-Statistik: $$ F = \frac{MS_B}{MS_W} = \frac{SS_B / (k-1)}{SS_W / (N-k)} $$
$F$ folgt einer F-Verteilung mit $ (k-1, N-k) $ Freiheitsgraden [1].

#### Zentrale Begriffe und Modelle

##### Hypothesentests
- **Nullhypothese** $H_0$: "kein Unterschied" – z.B. $H_0: \mu_1 = \mu_2$ [1]
- **Alternativhypothese** $H_1$: "Unterschied vorhanden" – z.B. $H_1: \mu_1 \neq \mu_2$ [1]
- **Signifikanzniveau** $\alpha = 0{,}05$: Fehler 1. Art (5%) [1]
- **p-Wert**: $p < \alpha$ → $H_0$ verwerfen [1]


##### t-Test
- **Anwendung**: 2 Gruppen (vorher/nachher) [1]
- **Annahmen**: Normalverteilt, unabhängig
- **Welch-t-Test**: Ungleiche Varianzen $s_X^2 \neq s_Y^2$ [1] [2]
- **Interpretation**: $p < 0{,}05$ → signifikant


##### ANOVA
- **Anwendung**: ≥3 Gruppen (Maschine A/B/C) [1]
- **Zerlegung**: $SS_{\text{gesamt}} = SS_B + SS_W$
- **F-Statistik**: $F = \frac{MS_B}{MS_W}$ [1]


##### Statistische Prozesskontrolle (SPC)
- **Zufällige Ursachen**: Normale Schwankung [1]
- **Spezielle Ursachen**: Maschine defekt [1]
- **Phase I/II**: Stabilisieren/Überwachen [1]

##### Control Charts
| Chart | Datenart | Überwacht | Center Line | Grenzen |
|-------|----------|-----------|-------------|---------|
| **X̄** | Stetig | Mittelwert | $\bar{\bar{X}}$ | $\bar{\bar{X}} \pm A_2 \bar{R}$ [1] |
| **R** | Stetig | Streuung | $\bar{R}$ | $D_3 \bar{R}$ bis $D_4 \bar{R}$ [1] |
| **p** | Attribut | Ausschuss | $\bar{p}$ | $\bar{p} \pm 3\sqrt{\frac{\bar{p}(1-\bar{p})}{n}}$ [1] |

## **Praktischer Teil (Code)**

#### Datenbeschreibung
Für die Durchführung der Analysen wurde ein eigenes Datenset mit zufallsgenerierten Daten erstellt. Dieses soll die Produktion von Werkstücken, genauer den Enddurchmesser gedrehter Rohlinge, repräsentieren. Dabei wurde der gewünschte Enddurchmesser als Medianwert gewählt und jeweils mit unterschiedlicher Standardabweichung um diesen Median 200 Werte zufallsgeneriert (gauss'sche Normalverteilung). Es wurden insgesamt für 3 Produktionschargen ('Batches') Daten generiert. Der Enddurchmesser/Medianwert wurde auf 100 mm gewählt. Die Standardabweichungen sind mit jedem 'Batch' absteigend gewählt worden: bei 'Batch 1' 6, bei 'Batch 2' 3 und bei 'Batch 3' 1, um eine vermeintliche Verbesserung des Produktionsprozesses pro Batch zu zeigen. Zusätzlich wurde eine Toleranz definiert (1 mm) und anhand dessen die 'Batch#_spec'-Zellen befüllt, was anzeigt, ob die gedrehten Werkstücke innerhalb der gewünschten Toleranz bearbeitet wurden. Liegt der Enddurchmesser zwischen 99 und 101 mm, ist 'Batch#_spec' = 1, ansonsten 0.

```python
# IMPORTS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# LOAD DATA
data = pd.read_csv("daten/Process.csv", sep=";")

batch1 = list(data['Batch1'])
batch2 = list(data['Batch2'])
batch3 = list(data['Batch3'])

# KLASSE IMPLEMENTIERT ANALYSEFUNKTIONEN
class DataParalysis:
    def TTest():
        """Führt den T-Test durch"""

        # 4. Unabhängiger t-Test (zwei Gruppen)
        t_stat, p_value = stats.ttest_ind(batch1, batch2, equal_var=False)  # Welch's t-Test

        print("\nT-Test Ergebnisse:")
        print(f"T-Statistik: {t_stat:.4f}")
        print(f"P-Wert: {p_value:.4f}")
        alpha = 0.05
        if p_value < alpha:
            print("Ergebnis: Signifikanter Unterschied (H0 abgelehnt)")
        else:
            print("Ergebnis: Kein signifikanter Unterschied")

        # 5. Visualisierung
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        plt.hist(batch1, alpha=0.7, label='Batch1', bins=20)
        plt.hist(batch2, alpha=0.7, label='Batch2', bins=20)
        plt.xlabel('Durchmesser')
        plt.ylabel('Messpunkte')
        plt.legend()
        plt.title('Verteilungen')

        plt.subplot(1, 2, 2)
        box_data = [batch1, batch2]
        plt.boxplot(box_data, labels=['Batch1', 'Batch2'])
        plt.ylabel('Durchmesser')
        plt.title('Boxplot-Vergleich')
        plt.tight_layout()
        plt.show()
    
    def Anova():
        """Führt ANOVA (Varianzanalyse) durch"""

        # Deskriptive Statistiken
        groups = [batch1, batch2, batch3]
        labels = ['Batch1', 'Batch2', 'Batch3']
        for i, g in enumerate(groups):
            print(f"{labels[i]}: Mittelwert = {np.mean(g):.3f}, Std = {np.std(g):.3f}")

        # Einweg-ANOVA (Vergleich mehrerer Gruppen)
        f_stat, p_value = stats.f_oneway(batch1, batch2, batch3)

        print("\nANOVA Ergebnisse:")
        print(f"F-Statistik: {f_stat:.4f}")
        print(f"P-Wert: {p_value:.4f}")
        alpha = 0.05
        if p_value < alpha:
            print("Ergebnis: Signifikanter Gesamtunterschied (H0 abgelehnt)")
        else:
            print("Ergebnis: Kein signifikanter Gesamtunterschied")

        # Post-hoc Test (Tukey HSD) falls ANOVA signifikant
        if p_value < alpha:
            # Daten für Tukey zusammenfassen
            data_tukey = np.concatenate(groups)
            groups_tukey = np.repeat([0, 1, 2], [len(g) for g in groups])
            tukey = pairwise_tukeyhsd(data_tukey, groups_tukey, alpha=0.05)
            print("\nTukey HSD Post-hoc Test:")
            print(tukey)

        plt.figure(figsize=(12, 4))

        plt.subplot(1, 2, 1)
        for i, g in enumerate(groups):
            plt.hist(g, alpha=0.7, label=labels[i], bins=15)
        plt.xlabel('Durchmesser')
        plt.ylabel('Messpunkte')
        plt.legend()
        plt.title('Verteilungen der Phasen')

        plt.subplot(1, 2, 2)
        plt.boxplot(groups, labels=[l[:8] for l in labels])  # Labels kürzen
        plt.ylabel('Durchmesser')
        plt.title('Boxplot-Vergleich (ANOVA)')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    
    def Charts():
        """Erstellt die Xbar-, R- und P-Charts"""

        # Xbar/R-Constants für n=3 (da du 3 Batches = 3 Werte pro Stichprobe hast)
        A2 = 1.023
        D3 = 0.0
        D4 = 2.574

        df = data

        # Prüfen, ob Toleranzspalten vorhanden sind
        # (Batch1_spec, Batch2_spec, Batch3_spec werden als 0/1 interpretiert)
        spec_cols = ["Batch1_spec", "Batch2_spec", "Batch3_spec"]
        batch_cols = ["Batch1", "Batch2", "Batch3"]

        # -----------------------------
        # Xbar- und R-Chart (über alle 3 Batches als Subgruppe)
        # -----------------------------
        # Subgruppen: jede Zeile ist eine Stichprobe mit n=3 Messwerten
        subgroup_values = df[batch_cols].values

        # Xbar je Stichprobe
        xbar = subgroup_values.mean(axis=1)
        # Range je Stichprobe
        R = subgroup_values.max(axis=1) - subgroup_values.min(axis=1)

        # Mittelwerte über alle Stichproben
        Xbar_bar = xbar.mean()
        R_bar = R.mean()

        # Regelgrenzen Xbar-Chart
        UCL_xbar = Xbar_bar + A2 * R_bar
        LCL_xbar = Xbar_bar - A2 * R_bar

        # Regelgrenzen R-Chart
        UCL_R = D4 * R_bar
        LCL_R = D3 * R_bar  # bei n=3 ist D3 = 0

        # ----- Plot Xbar-Chart -----
        plt.figure(figsize=(10, 4))
        plt.plot(xbar, marker='o', linestyle='-', label='Stichprobenmittelwerte')
        plt.axhline(Xbar_bar, color='green', linestyle='--', label='CL (X̄̄)')
        plt.axhline(UCL_xbar, color='red', linestyle='--', label='UCL')
        plt.axhline(LCL_xbar, color='red', linestyle='--', label='LCL')
        plt.axhline(101, color='blue', linestyle=':', label='Toleranz')
        plt.axhline(99, color='blue', linestyle=':')
        plt.title('X̄-Chart')
        plt.xlabel('Datenpunkte')
        plt.ylabel('Durchmesser [mm]')
        plt.legend()
        plt.grid(True)

        # ----- Plot R-Chart -----
        plt.figure(figsize=(10, 4))
        plt.plot(R, marker='o', linestyle='-', label='Range je Stichprobe')
        plt.axhline(R_bar, color='green', linestyle='--', label='CL (R̄)')
        plt.axhline(UCL_R, color='red', linestyle='--', label='UCL')
        plt.axhline(LCL_R, color='red', linestyle='--', label='LCL')
        plt.title('R-Chart')
        plt.xlabel('Datenpunkte')
        plt.ylabel('Range [mm]')
        plt.legend()
        plt.grid(True)

        # ----- Plot P-Chart -----
        # Batch*_spec = 1 -> innerhalb ±1 mm, 0 -> außerhalb
        # pro Stichprobe: 3 Teile (Batch1/2/3), also n_i = 3
        spec_matrix = df[spec_cols].values
        n_i = spec_matrix.shape[1]
        p_i = 1.0 - spec_matrix.mean(axis=1)  # Anteil außerhalb Spezifikation pro Stichprobe

        p_bar = p_i.mean()

        # Regelgrenzen P-Chart (konstantes n_i = 3)
        sigma_p = np.sqrt(p_bar * (1 - p_bar) / n_i)
        UCL_p = p_bar + 3 * sigma_p
        LCL_p = p_bar - 3 * sigma_p
        LCL_p = max(LCL_p, 0.0)  # LCL nicht kleiner 0

        plt.figure(figsize=(10, 4))
        plt.plot(p_i, marker='o', linestyle='-', label='Ausschussanteil pro Stichprobe')
        plt.axhline(p_bar, color='green', linestyle='--', label='CL (p̄)')
        plt.axhline(UCL_p, color='red', linestyle='--', label='UCL')
        plt.axhline(LCL_p, color='red', linestyle='--', label='LCL')
        plt.title('P-Chart')
        plt.xlabel('Datenpunkte')
        plt.ylabel('Ausschussanteil')
        plt.ylim(bottom=0)
        plt.legend()
        plt.grid(True)

        plt.tight_layout()
        plt.show()
```

```python
DataParalysis.TTest()
DataParalysis.Anova()
DataParalysis.Charts()
```

#### Interpretation
#### Boxplot Interpretation:
Der Boxplot hat an seinen Enden flache Linien parallel zur X-Achse; diese zeigen jeweils das Minimum und Maximum der Werte an. Sind Punkte im Diagramm zu erkennen, bedeutet dies starke Ausreißer eines Datenpunktes im Datensatz. Die orange Linie in der Mitte zeigt immer den Medianwert an. Die Box selbst visualisiert die jeweiligen Quantile: Der untere Boxrand zeigt das 1. Quantil und der obere Boxrand das 3. Quantil.

#### T-Test
Der t-Test zeigt, wie sich die Durchmesser zwischen Produktionscharge 1 und 2 verbessert haben. In Batch 1 ist die Grafik breit und zeigt eine hohe Streuung auf, während in Batch 2 sich diese Streuung schon stark reduziert und näher an den gewünschten Enddurchmesser von 100 mm annähert. Im danebenliegenden Boxplot-Vergleich sieht man, wie nicht nur der Median (orange Linie) etwas näher an die 100 rückt, sondern auch wie insgesamt die maximalen und minimalen Werte (die flachen Balken jeweils an den Enden des Plots) kleiner werden.

#### ANOVA
Die ANOVA (auch Varianzanalyse), bekannt dafür, vergleicht alle 3 Batches und zeigt uns ähnlich wie der t-Test zuvor, dass der Mittelwert mit jedem Batch und jeder Produktionsverbesserung näher an den Wunschdurchmesser kommt. Außerdem sehen wir im Boxplot, dass die Minima und Maxima ebenfalls näher an die 100 mm kommen und immer besser auch innerhalb unserer Toleranz liegen.

#### Charts / Regelkarten
#### XBar-Chart
Der X̄-Chart zeigt uns, ob unser Prozess stabil oder eher instabil wirkt. Ein stabiler Prozess würde um einen Wert schwingen, während ein instabiler Prozess einen Trend aufzeigen würde, in welchem die Werte immer größer oder kleiner würden. Im Endeffekt beantwortet das X̄-Chart, ob sich der Mittelwert des Prozesses über die Zeit verschiebt. Eingezeichnet sind der Median (CL, auch Center Line genannt), die jeweiligen Eingriffsgrenzen (UCL & LCL) sowie die Spezifikationsgrenzen (die Toleranz zwischen 99 und 101 mm). Ebenfalls sind noch die Mittelwerte der Datenpunkte eingezeichnet. Mittelwerte daher, da die Zeile der Datenpunkte (die 3 Werte für jedes Batch) gemittelt werden muss und dann als ein Datenpunkt eingezeichnet wird. Man sieht hier, dass der Prozess relativ stabil ist, da er um den Median von 100 mm schwingt und kein langfristiger Drift erkennbar ist. UCL und LCL können entweder als 3-Sigma zum Median berechnet werden oder aus einer Standardtabelle genommen werden; hier wurde die Standardtabelle verwendet.


#### R-Chart
Das R-Chart zeigt uns die Spannweite innerhalb der gemittelten Werte von X̄. Die Spannweite wird folgendermaßen berechnet: $R = \max - \min$. Im Endeffekt beantwortet das R-Chart, ob sich die Streuung (Breite der Verteilung) zwischen den Teilstichproben über die Zeit ändert. Eingezeichnet sind ebenfalls die Eingriffsgrenzen (UCL & LCL) sowie der Medianwert (CL). In unserem Chart schwankt zwar die Spannweite relativ stark, bleibt aber grob in einem Band und zeigt ebenfalls keinen Drift in eine größer- oder kleinerwerdende Richtung. Für den betrachteten Prozess ist die Schwankung jedoch viel zu hoch und liegt deutlich über der gewünschten Toleranz von 1 mm.


#### P-Chart
Das p-Chart zeigt uns den Ausschussanteil der Werkstücke an. Da im Datenset nur 1 oder 0 (innerhalb der Toleranz/außerhalb der Toleranz) berücksichtigt wurde, ist der Graph zickzackförmig. Obwohl das Argument vorgebracht werden könnte, dass Bauteile nach dem Drehen, die noch über 101 mm haben, nicht direkt Ausschuss sind, sondern wieder auf die Drehbank gebracht und nachbearbeitet werden könnten, wurde diese Möglichkeit im Datenset jedoch nicht berücksichtigt. Daher fallen die meisten Werkstücke am Ende durch und erreichen die geforderte Toleranz nicht.

## **Diskussion / Fazit**

#### Bewertung der Methode, Grenzen, Ausblick

Insgesamt sind die Methoden der statistischen Prozesskontrolle (SPC) relativ hilfreiche Methoden, um einen Prozess zu überwachen. Besonders hervorzuheben sind hier der t-Test und die ANOVA (Varianzanalyse). Sie erlauben einen tiefen Einblick in die Qualität eines Prozesses und helfen, frühzeitig Fehler festzustellen, zu agieren oder einen fehlerhaften Prozess zu verbessern.
Jedoch sind Control Charts bei kleinen Stichproben (n<5) weniger aussagekräftig, reagieren verzögert auf kleine Verschiebungen und können durch Zufalls-Muster falsche Alarme auslösen.
